```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
A = x_axis
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
B = y_axis
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(A, B)
```

```python
# the "fig, ax = plt.subplots()" line of code can also be written like this:

# Create the plot with "ax.plt()"
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(A, B)

# Note: To change figure-level attributs (such as axes labels, a title, or a legend)
# or save figure as an image, then use "fig = plt.figure()"


# Use "plt.show()" - this function looks for all active objects and opens a window
# that displays them if you have two or more sets of data to plot.

# IMPORTANT: "plt.show()" should only be used once per session because it can cause
# data to display improperly. IF you have problems you may need to restart your kernal
# and clean the output.
#  kernal --> Restart & Clean output
```

```python
# Mod 5.1.4 Annotate Charts - with MatPlotLib, we can labels to the axes, a title, and a legend.

# Annotate Charts using the MatPlotLib Method

# Create the plot and add a label for the legend
plt.plot(A, B, label = "Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare$")
# Set the y-limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a legend
plt.legend()
```

```python
# We can also change the width and color of the line and add markers for the data points.
# This is done by declaring parameters in the "plt.plot()" 
# To declare the parameters for the line color, width, and marker, we use color=, width(),
# and marker= inside the plt.plot() function.

# Create the plot.
plt.plot(A, B, marker="*", color="purple", linewidth=3, label="Atlanta")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare$")
# Set the y-limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
plt.legend()
```

```python
# Annotate Charts using the Object-Oriented Interface
# fig, ax=plt.subplots(figsize=(w, h))
# ax.plot(x, y, label="line")
# ax.set_ylim(min, max)
# ax.set_xlim(min, max)
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')
# ax.set_title("Title")
# ax.legend()
# ax.grid()
# plt.savfig(" add a path and figure ext

fig, ax = plt.subplots(figsize= (15,10))
ax.plot(A, B, label = 'Austin', color="orange", linewidth=2, marker="*")
ax.set_ylim(0, 45)
ax.set_xlabel('Date')
ax.set_ylabel("Fare$")
ax.set_title("PyBer $ by Month")
ax.legend()
ax.grid()
ax.text(0.5, 1.0, "PyBer $ by Month")
#  plt.savfig("add a path and figure extension")
```

```python
# Mod 5.1.7 Create Scatter Plots and Bubble Charts.

# A scatter plot visually depicts a positive, negative, or no-correlation 
# relationshop between two factors.

# Creating a Scatter Plot using the MatLab Method.

# plt.scatter(x_axis, y_axis, s=y_axis)

# To change the size of our marker on the graph, we have to use the "s" or size parameter, for the marker.
# The "s" parameter must equal to a feature of our data. 
# After running this cell, the marker size has changed based on the value of the fare in USD.

# some of the markers look very small. Let's make them bigger so that the chart is more legible 
# and impactful. We can adjust the size by multiplying the data in the y_axis by any value.

# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data * 3)
# plt.scatter(x_axis, y_axis, s=y_axis_larger)

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data * 10)
plt.scatter(y_axis, x_axis, s=y_axis_larger, color="skyblue", label="Boston", edgecolors="black" )
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.xlim(0, 50)
plt.grid()
plt.legend()
```

```python
# Instead of using the 'for' loop to create the list y_axis_larger for the size of the parameter of the
# marker, we can use Python's list comprehension technique inside the plt.scatter() function.
# new_list = [expression for the item in list if conditional]

# plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

plt.scatter(x_axis, y_axis, s = [i * 15 for i in y_axis], color="skyblue", label="Boston", edgecolors="black")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.ylim(0, 50)
plt.title("PyBer Fare by Month")
plt.legend()
plt.grid()
```

```python
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=8)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=8)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=8)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=8)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
